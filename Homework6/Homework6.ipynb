```python
import numpy as np
import matplotlib.pyplot as plt
from GraphCutActiveContours import GraphCutActiveContours
import cv2
import imageio
```

```python
img_simple = np.zeros((64,64))
img_simple[25:31,25:31] = 1

c0_simple = np.zeros((64,64))
c0_simple[19,19:35] = 1
c0_simple[35,19:35] = 1
c0_simple[19:35,19] = 1
c0_simple[19:35,35] = 1

plt.figure()
plt.subplot(1,2,1)
plt.imshow(img_simple, cmap='gray')
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(c0_simple, cmap='gray')
plt.title("initial contour")
plt.show()


seg = GraphCutActiveContours(img_simple)
res = seg.segmentation(c0_simple)

plt.subplot(1,2,1)
plt.imshow(img_simple, cmap='gray')
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(res, cmap='gray')
plt.title("final contour")
plt.show()
```
