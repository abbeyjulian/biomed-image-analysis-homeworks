A. Import darknet framework, load YOLO models, and prepare list of images

```python
import matplotlib.pyplot as plt
from darknetpy.detector import Detector
import os
import time
import sys
import json

# darknet usage: https://pypi.org/project/darknetpy/
# yolo usage: https://pjreddie.com/darknet/yolo/
```

```python
path = os.getcwd()
```

```python
# load models

detector = Detector(path+'/cfg/voc.data',
                    path+'/cfg/yolov3.cfg',
                    path+'/yolov3.weights')

tiny = Detector(path+'/cfg/voc.data',
                    path+'/cfg/yolov3-tiny.cfg',
                    path+'/yolov3-tiny.weights')
```

```python
# load VOC validation images

images = []

f = open(path+"/voc/TrainVal/2011_val.txt", "r")
for im in f:
    images.append(im[0:-1]) # remove extra newline
f.close()

print(len(images))
```

B. Predict bounding boxes on images

```python
# standard model
results_std = [] # list of dicts {res: [results list], time: execution_time}

for k in range(5495, len(images)):
    t1 = time.time()
    res = detector.detect(images[k])
    t = time.time() - t1
    result = {'im': k, 'res': res, 'time': t}
    results_std.append(result)
    print("image %d predicted in time %1.4f" % (k, t))
    sys.stdout.flush() # force to be seen
    # save result (to allow stop & restart of predictions)
    saved_results = open('results/save_res_std_%d.json' % (k), 'w', encoding='utf-8')
    json.dump(result, saved_results) 
    saved_results.write("\n")
    saved_results.close()
```

```python
# tiny model
results_tiny = [] # list of dicts {res: [results list], time: execution_time}

for k in range(4043, len(images)):
    t1 = time.time()
    res = tiny.detect(images[k])
    t = time.time() - t1
    result = {'im': k, 'res': res, 'time': t}
    results_tiny.append(result)
    print("image %d predicted in time %1.4f" % (k, t))
    sys.stdout.flush() # force to be seen
    # save result (to allow stop & restart of predictions)
    saved_results = open('results/save_res_tiny_%d.json' % (k), 'w', encoding='utf-8')
    json.dump(result, saved_results) 
    saved_results.write("\n")
    saved_results.close()
```

C. Load ground-truth bounding box labels

```python
# borrowed from darknet voc_label.py
import xml.etree.ElementTree as ET
from os import listdir, getcwd
from os.path import join

classes = ["aeroplane", "bicycle", "bird", "boat", "bottle", "bus", "car", "cat", "chair", "cow", "diningtable", "dog", "horse", "motorbike", "person", "pottedplant", "sheep", "sofa", "train", "tvmonitor"]

def convert_annotation(year, image_id, k):
    in_file = open(path+'/voc/TrainVal/VOCdevkit/VOC%s/Annotations/%s.xml'%(year, image_id))
    res = []
    tree=ET.parse(in_file)
    root = tree.getroot()
    size = root.find('size')
    w = int(size.find('width').text)
    h = int(size.find('height').text)

    for obj in root.iter('object'):
        difficult = obj.find('difficult').text
        cls = obj.find('name').text
        if cls not in classes or int(difficult)==1:
            continue
        xmlbox = obj.find('bndbox')
        b = (float(xmlbox.find('xmin').text), float(xmlbox.find('xmax').text), float(xmlbox.find('ymin').text), float(xmlbox.find('ymax').text))
        res.append({'class': cls, 'left': b[0], 'right': b[1], 'bottom': b[2], 'top': b[3]})
    
    annotation = {'im': k, 'res': res}
    return annotation

year = '2011'
ground_truth = []
image_ids = open(path+'/voc/TrainVal/VOCdevkit/VOC%s/ImageSets/Main/%s.txt'%(year, 'val')).read().strip().split()
for k in range(len(image_ids)):
    ground_truth.append(convert_annotation(year, image_ids[k], k))

print(len(ground_truth))
```

D. Define IoU and mAP metrics

```python
def IoU(box1, box2):
    # find overlapping region
    overlap_x1 = max(box1['left'], box2['left'])
    overlap_y1 = max(box1['top'], box2['top'])
    overlap_x2 = min(box1['right'], box2['right'])
    overlap_y2 = min(box1['bottom'], box2['bottom'])
    intersection = max(0,((overlap_x2-overlap_x1)*(overlap_y2-overlap_y1)))
    
    area_box1 = (box1['right']-box1['left'])*(box1['bottom']-box1['top'])
    area_box2 = (box2['right']-box2['left'])*(box2['bottom']-box2['top'])
    
    union = area_box1 + area_box2 - intersection
    
    return intersection / union

def get_precision(stat):
    if (stat['tp']==0):
        return 0
    return stat['tp']/(stat['tp']+stat['fp'])

def get_recall(stat):
    if (stat['tp']==0):
        return 0
    return stat['tp']/(stat['tp']+stat['fn'])

def get_p_interp(stat, r):
    p_interp = 0
    for i in range(len(stat['precision'])):
        if stat['recall'][i] >= r and stat['precision'][i] >= p_interp:
            p_interp = stat['precision'][i]
    
    return p_interp

def mAP(pred, truth):
    conf_thresh = 0.75
    iou_thresh = 0.5
        
    classes = ["aeroplane", "bicycle", "bird", "boat", "bottle", "bus", "car", "cat", "chair", "cow", "diningtable", "dog", "horse", "motorbike", "person", "pottedplant", "sheep", "sofa", "train", "tvmonitor"]
    stats = {}
    preds = {}
    nums = {}
    AP = {}
    for c in classes:
        nums[c] = {'tp': 0, 'fp': 0, 'fn': 0}
        stats[c] = {'precision': [], 'recall': [], 'p_interp': []}
        preds[c] = []
        AP[c] = 0
    
    # separate predictions in classes and then sort on probability
    for k in range(len(pred)):
        im_pred = pred[k]
        for i in range(len(im_pred['res'])):
            cls = im_pred['res'][i]['class']
            data = im_pred['res'][i]
            data['im'] = k
            preds[im_pred['res'][i]['class']].append(data)
    for c in classes:
        preds[c] = sorted(preds[c], key = lambda i: i['prob'],reverse=True)
        
    # for each class, build precision and recall lists
    mAP = 0
    for c in classes:
        used = [] # keep track of matched ground-truths to avoid duplicates; (image_id, box_id)
        for p in preds[c]:
            if p['prob'] > conf_thresh:
                iou = 0
                true = None
                im_truth = truth[p['im']]
                for t in range(len(im_truth['res'])):
                    if im_truth['res'][t]['class'] == c:
                        iou_new = IoU(p, im_truth['res'][t])
                        if iou_new > iou:
                            iou = iou_new
                            true = (p['im'], t)
                if iou < iou_thresh or true in used: # false positive - iou too small or duplicate
                    nums[c]['fp'] += 1
                elif true is None: # false negative - couldn't match to a ground truth box
                    nums[c]['fn'] += 1
                else: # true positive - got it!
                    nums[c]['tp'] += 1
                    used.append(true)
                    print(p)
                stats[c]['precision'].append(get_precision(nums[c]))
                stats[c]['recall'].append(get_recall(nums[c]))
                stats[c]['p_interp'].append(get_p_interp(stats[c], get_recall(nums[c])))
        # check for missed detections (more false negatives)
        for i in range(len(truth)):
            for j in range(len(truth[i]['res'])):
                if truth[i]['res'][j]['class'] == c and (i,j) not in used:
                    nums[c]['fn'] += 1
                    stats[c]['precision'].append(get_precision(nums[c]))
                    stats[c]['recall'].append(get_recall(nums[c]))
                    stats[c]['p_interp'].append(get_p_interp(stats[c], get_recall(nums[c])))
        
        # print(stats[c])
        for i in range(len(stats[c]['precision'])-1):
            AP[c] += (stats[c]['recall'][i+1]-stats[c]['recall'][i])*stats[c]['p_interp'][i+1]
        print("average precision class %s = %1.4f" % (c, AP[c]))
        mAP += AP[c]

    mAP = mAP / len(classes)
    return mAP
```

E. Calculate mAP; save metrics and execution time

```python
# load full results
results_std = []
for file_name in [file for file in os.listdir('results/') if 'std' in file]:
    with open('results/' + file_name) as json_file:
        results_std.append(json.load(json_file))
print(len(results_std))

results_tiny = []
for file_name in [file for file in os.listdir('results/') if 'tiny' in file]:
    with open('results/' + file_name) as json_file:
        results_tiny.append(json.load(json_file))
print(len(results_tiny))
```

```python
# calculate mAP
mAP_std = mAP(results_std, ground_truth)
mAP_tiny = mAP(results_tiny, ground_truth)

print(mAP_std)
print(mAP_tiny)
```
